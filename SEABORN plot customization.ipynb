```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('covid.csv')
df.head()
```

```python
r1=df['month'].value_counts().reset_index()
r1
```

```python
sns.barplot(data=r1,x='index',y='month')
```

```python
sns.barplot(data=r1,x='index',y='month')
plt.title('count of months',fontsize=14,color='blue')
plt.xlabel('months')
plt.ylabel('count')
plt.show()
```

```python
print(len(df['popData2018'].unique()))
print(len(df['countries'].unique()))
```

```python
data=df[['countries','popData2018']]
data.duplicated().sum()
```

```python
data=data.drop_duplicates()
```

```python
print(data.shape)
```

```python
top7_pop=data.sort_values('popData2018',ascending=False).head()
top7_pop
```

```python
sns.set(rc={'figure.figsize':(5,4)})
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("paper")
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("talk")
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("notebook")
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style('dark')
sns.set_context("notebook")
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(5,4)})
sns.set_style('whitegrid')
sns.set_context("talk")
sns.set_context('poster')
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_context("poster", font_scale= .8, rc={"grid.linewidth":0.7})
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.xticks(rotation=90)
plt.show()
```

```python
palette=sns.color_palette("bright")
sns.palplot(palette)
```

```python
pal2=sns.color_palette("Set1")
sns.palplot(pal2)
```

```python
pal3=sns.color_palette("coolwarm")
sns.palplot(pal3)
```

```python
sns.set_palette("RdBu",10)
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.ylabel('population data 2018')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("talk")
sns.despine()
sns.barplot(data=top7_pop,x='countries',y='popData2018')
plt.title('country wise population')
plt.xticks(rotation=90)
plt.show()
```
