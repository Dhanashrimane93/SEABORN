```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df_1=pd.read_csv('Healthexp.csv')
```

```python
# Q.1 CREATE a line plot ofcountry wise spending on helathcare and apply the following customization
df_2=df_1.groupby(['Country']).mean().reset_index()
```

```python
df_2[['Spending_USD','Life_Expectancy']]=df_2[['Spending_USD','Life_Expectancy']].astype(int)
```

```python
df_2=df_2.drop(['Year'],axis=1)
df_2
```

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.barplot(data=df_2,x='Country',y='Spending_USD')
plt.title('Contry wise Spending (USD) on healthcare')
plt.xlabel('Country')
plt.ylabel('Spending(USD)on healthcare')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q.2 cerate multiple plots using fascetgrid() function and apply the following customizationto the plot

df_2015=df_1.groupby(['Country','Year']).mean().reset_index()
```

```python
df_2015[['Spending_USD','Life_Expectancy']]=df_2015[['Spending_USD','Life_Expectancy']].astype(int)
```

```python
df_2015=df_2015[df_2015["Year"]>2015]
df_2015.head()
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
sea=sns.FacetGrid(df_2015,col="Year",col_wrap=3,height=5,aspect=1.2)
sea.map(sns.scatterplot,"Country","Spending_USD")
plt.show()
```
