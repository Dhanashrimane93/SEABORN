```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('covid.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['qtr']=df['date'].dt.quarter
df.head()
```

```python
df['qtr'].value_counts()
```

```python
df.columns
```

```python
# strip plot
```

```python
# plot strip plot representing cases for india belgium russia

df_3c=df[df['countries'].isin(['India','Belgium','Russia'])]
df_3c.shape
```

```python
df_3c['countries'].value_counts()
```

```python
sns.stripplot(data=df_3c,x='countries',y='cases')
plt.show()
```

```python
# plot strip plot
```

```python
sns.stripplot(data=df_3c,x='countries',y='cases',hue='month')
plt.title('country wise cases -strippolt')
plt.show()
```

```python
# swarm plot
```

```python
sns.swarmplot(data=df_3c,x='countries',y='cases')
plt.title('country wise covid cases-swarmplot')
plt.show()
```

```python
sns.swarmplot(data=df_3c,x='countries',y='cases',hue='month')
plt.title('country wise covid case -swarmplot')
plt.show()
```

```python
#displot
```

```python
df_3c_mar_apr=df_3c[df_3c['month'].isin([3,4])]
df_3c_mar_apr.shape
```

```python
sns.displot(data=df_3c_mar_apr,x='cases')
plt.title('displot of cases fot india belgium,russia')
plt.show()
```

```python
sns.displot(data=df_3c_mar_apr,x='cases',kind='kde')
plt.show()
```

```python
# countplot
```

```python
sns.countplot(data=df,x='month')
plt.title('count plot for month')
plt.show()
```

```python
sns.countplot(data=df_3c_mar_apr,x='month',hue='countries')
plt.title('count plot for month')
plt.show()
```
