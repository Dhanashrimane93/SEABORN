```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df_1=pd.read_csv('Healthexp.csv')
```

```python
# Q.1 create SCATTER PLOT OF YEAR WISE LIFE expectancy based on countries and apply ath customization

sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.scatterplot(data=df_1,x='Year',y='Life_Expectancy',hue='Country',style='Country',size='Country')
plt.title('year wise life expectancy based on countries')
plt.xlabel('Year')
plt.ylabel(' life expectancy over year')
plt.legend(bbox_to_anchor=(1,1))
plt.grid()
plt.show()
```

```python
# Q.2 CREATE BOXPLOT OF COUNTRY WISE SPENDING IN USD Boxplot and apply folowing custumization


sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("hls")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.boxplot(data=df_1,x='Country',y='Spending_USD')
plt.title('country wise spending in usd -boxplot')
plt.xlabel('country')
plt.ylabel('spending (usd) on healthcare')
plt.grid()
plt.show()
```

```python
#Q .3 create a violin plot of country wise life expectancy violin plot and apply the customization

sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("husl")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.violinplot(data=df_1,x='Country',y='Life_Expectancy')
plt.title('country wise life expectancy-violin plot')
plt.xlabel('country')
plt.ylabel('life expectanct')
plt.show()
```
