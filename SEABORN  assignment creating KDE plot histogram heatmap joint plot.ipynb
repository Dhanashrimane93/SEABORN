```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df_1=pd.read_csv('Healthexp.csv')
```

```python
# Q.1 create kde plot for spending on healthcare based on countries and apply the 
# following customization
```

```python
sns.set(rc={'figure.figsize':(15,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.kdeplot(data=df_1,x='Spending_USD',hue='Country',palette='tab10')
plt.title('Kde plot for spending on healthcare based on countries')
plt.show()
```

```python
# Q.2 CREATE HISTOGRAM of histogram for life expectancy based on year (2019,2020,2021) and apply customization
```

```python
df_2018=df_1[df_1["Year"]>2018]
df_2018.head()
sns.set(rc={'figure.figsize':(15,8)})
sns.set_palette("hls")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.histplot(data=df_2018,x='Life_Expectancy',hue='Year',palette='bright')
plt.title('histogram for life expectancy based on year 2019,2020,2021')
plt.show()
```

```python
#Q.3 create heatmap apply the following customization
```

```python
corr=df_1.corr()
corr
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(data=corr,annot=True,cmap='RdBu')
plt.show()
```

```python
# Q.4 creat a joint plot and apply customiztaion
```

```python
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.jointplot(data=df_1,x='Spending_USD',y='Year',hue='Country',palette='bright',
              kind='kde',height=10)
plt.show()
```
