```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import  matplotlib.pyplot as plt
```

```python
df_1=pd.read_csv('Healthexp.csv')
```

```python
#Q.1 create strip plot of country wise spending inusd strip plot and apply the following customization

sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.stripplot(data=df_1,x='Country',y='Spending_USD')
plt.title('country wise spending in usd -strip plot')
plt.xlabel('country')
plt.ylabel('spending (usd) on healthcare')
plt.grid()
plt.show()
```

```python
#Q.2 CREATE swarm plot of country wise life expectancy based on year and apply the following custom

sns.set(rc={'figure.figsize':(12,8)})
sns.set_palette("hls")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.swarmplot(data=df_1,x='Country',y='Life_Expectancy',hue='Year')
plt.title('country wise life expectancy based on year')
plt.xlabel('country')
plt.ylabel('country wise life expectancy based on year')
plt.legend(ncol=2,fontsize=12,markerscale=0.8,bbox_to_anchor=(1,1))
plt.grid()
plt.show()
```

```python
#Q.3 create displot of diplot of spending on health in usd and apply the following custoomization

sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.displot(data=df_1,x='Spending_USD')
plt.title('DISPLOT SPENDING ON HEALTHCARE (IN USD)')
plt.xlabel('Spending on healthcare in USD')
plt.show()
```

```python
# Q.4 create countplot of countplot for country and apply the customization

sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper",font_scale=2,rc={"grid.linewidth":0.5})
sns.countplot(data=df_1,x='Country')
plt.title('countplot for country')
plt.xlabel('country')
plt.grid()
plt.show()
```
