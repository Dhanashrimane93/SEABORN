```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df_1=pd.read_csv('Healthexp.csv')
```

```python
#Q.1 create a lineplot of mean of life expectancy based on year and country and apply the following customization

df_2=df_1.groupby(['Country','Year']).mean().reset_index()
df_2
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df_2,x='Year',y='Life_Expectancy',hue='Country',ci=None,style='Country',markers=True)
plt.title('MEan of Life_Expectancy based on years and country')
plt.grid()
plt.show()
```

```python
#Q.2  crceate bar plot of mean of spending usd based on year >2015
#and country and apply customization

df_2015=df_2[df_2["Year"]>2015]
df_2015.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=df_2015,x='Country',y='Spending_USD',hue='Year',ci=None)
plt.title('Mean of spending usd based on year(>2015) and country')
plt.grid()
plt.show()
```
